# 🔧 Preprocessing des Données
**Projet : Customer Churn Analytics**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../data/churn_data.csv')
print('Shape initial:', df.shape)
df.head()
```

## 1. Nettoyage des données

```python
# Convertir TotalCharges en numérique
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Supprimer les valeurs manquantes
print('Valeurs manquantes avant:', df.isnull().sum().sum())
df = df.dropna()
print('Valeurs manquantes après:', df.isnull().sum().sum())

# Supprimer customerID (inutile pour le ML)
df = df.drop(columns=['customerID'])
print('Shape après nettoyage:', df.shape)
```

## 2. Encodage des variables binaires

```python
# Encoder Yes/No en 1/0
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Encoder gender
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})

print('Encodage binaire terminé ')
df.head()
```

## 3. Encodage des variables catégorielles

```python
# One-Hot Encoding pour les colonnes multi-catégories
cat_cols = ['MultipleLines','InternetService','OnlineSecurity','OnlineBackup',
            'DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
            'Contract','PaymentMethod']

df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
print('Shape après encodage:', df.shape)
df.head()
```

## 4. Sauvegarde du dataset propre

```python
df.to_csv('../../data/churn_data_clean.csv', index=False)
print('Dataset propre sauvegardé ')
print('Shape final:', df.shape)
```
