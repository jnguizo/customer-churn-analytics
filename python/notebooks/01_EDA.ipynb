#  Analyse Exploratoire des Données (EDA)
**Projet : Customer Churn Analytics**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
df = pd.read_csv('../../data/churn_data.csv')
print('Shape:', df.shape)
df.head()
```

## 1. Aperçu général

```python
print('--- Infos générales ---')
df.info()
print('\n--- Valeurs manquantes ---')
print(df.isnull().sum())
print('\n--- Statistiques ---')
df.describe()
```

## 2. Taux de Churn

```python
churn_counts = df['Churn'].value_counts()
print(churn_counts)
print(f'\nTaux de churn : {churn_counts["Yes"] / len(df) * 100:.2f}%')

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Distribution du Churn')
plt.savefig('../../python/outputs/churn_distribution.png')
plt.show()
```

## 3. Churn par type de contrat

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title('Churn par type de contrat')
plt.savefig('../../python/outputs/churn_par_contrat.png')
plt.show()
```

## 4. Churn par ancienneté (tenure)

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, palette='Set2')
plt.title('Distribution du tenure par Churn')
plt.savefig('../../python/outputs/churn_par_tenure.png')
plt.show()
```

## 5. Corrélation des variables numériques

```python
df_num = df[['SeniorCitizen','tenure','MonthlyCharges','TotalCharges']].copy()
df_num['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

plt.figure(figsize=(8,6))
sns.heatmap(df_num.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de corrélation')
plt.savefig('../../python/outputs/correlation_matrix.png')
plt.show()
```
